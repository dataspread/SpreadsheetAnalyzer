```python
import pandas as pd
```

```python
def preprocess_df(df):
    column_name = [
        'File name', 'Formulae', 'Vertices', 'Edges', 'TACO Vertices', 'TACO Edges',
        'RR-Chain-Comp', 'RR-Chain-NoComp', 'RR-Comp', 'RR-NoComp', 'RF-Comp', 'RF-NoComp', 'FR-Comp', 'FR-NoComp', 'FF-Comp', 'FF-NoComp',
        'RR-GapOne', 'RR-GapOne-NoComp', 'RR-GapTwo', 'RR-GapTwo-NoComp', 'RR-GapThree', 'RR-GapThree-NoComp', 'RR-GapFour', 'RR-GapFour-NoComp', 
        'RR-GapFive', 'RR-GapFive-NoComp', 'RR-GapSix', 'RR-GapSix-NoComp', 'RR-GapSeven', 'RR-GapSeven-NoComp', 'NoType', 'NoType-NoComp'
    ]
    df.columns = column_name
    df.set_index(['File name'], inplace=True)
    return df
```

```python
df = pd.read_csv("./short_test_stat.csv", header=None)
df_new = pd.read_csv("./new_short_test_stat.csv", header=None)
df = preprocess_df(df)
df_new = preprocess_df(df_new)
```
